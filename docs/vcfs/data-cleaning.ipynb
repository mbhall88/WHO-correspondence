This notebook is for generating a list of all the filepaths on the EBI cluster for WHO VCFs we want.

```python
import pandas as pd
from pathlib import Path
from collections import defaultdict
```

```python
ebi_df = pd.read_csv("ena_tb_samples_clockwork_output.csv")
```

```python
who_runs = set(who_df["run"])
len(who_runs)
who_samples = set(who_df["biosample"]).union(set(who_df["sample"]))
```

```python
ebi_df_subset = ebi_df.query("`Seqrep.ena_run_accession` in @who_runs")
```

```python
dir2vcf = defaultdict(list)
with open("ena_minos.fofn") as fp:
    for line in map(str.strip, fp):
        p = Path(line)
        assert p.parts[14] == "Pipelines", line
        d = Path(p.parts[0]).joinpath(*p.parts[1:14])
        if "minos" not in line:
            continue
        dir2vcf[d].append(Path(p.parts[14]).joinpath(*p.parts[15:]))
```

```python
len(dir2vcf)
```

```python
minos_vcfs = []
for _, row in ebi_df_subset.iterrows():
    pipeline = Path(row["Pipeline_output"])
    vcfs = dir2vcf.get(pipeline)
    if not vcfs:
        minos_vcfs.append(None)
        continue
    
    if len(vcfs) == 1:
        minos_vcfs.append(vcfs[0])
        continue
        
    max_isolate = ""
    isolate_num = 0
    for v in vcfs:
        iso = v.parts[1]
        if "_" in iso:
            max_isolate = iso
            break
        else:
            if int(iso) > isolate_num:
                max_isolate = iso
                isolate_num = int(iso)
    
    max_min_ver = 0
    latest_vcf = None
    for v in vcfs:
        iso = v.parts[1]
        if iso != max_isolate:
            continue
        minor_ver = int(v.parts[3].split(".")[1])
        if minor_ver > max_min_ver:
            latest_vcf = v
    
    minos_vcfs.append(latest_vcf)
```

```python
ebi_df_subset["vcf"] = [str(p) for p in minos_vcfs]
```

```python
vcfs = []
for _, row in df.iterrows():
    root = Path(row["Pipeline_output"])
    v = Path(row["vcf"])
    p = root / v
    assert p.parts[8] == "Pipeline_root"
    vcfs.append(Path(p.parts[8]).joinpath(*p.parts[9:]))
```
